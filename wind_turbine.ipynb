```python
from fuzzywuzzy import process

company_list = ['Amazon Wind','Arbor Hill', 'Acciona','AES','Beebe','Black Hills','Blackstone','Blue Canyon Windpower',
                'Blue Summit','Brady Wind','Cassia','Cedar Springs Wind','CHI','Cimarron','Consolidated Edison',
                'Creston Ridge','DTE','Duke Energy','Echo','EDF Renewable','Enel','Eurus Combine Hills','Evergy',
                'Fluvanna Wind Energy','Foundation CA Fund','FPL Energy','Golden Hills','Green Pastures','Harvest',
                'Heartland','Hidalgo Wind Farm','Iowa Lakes','Lincoln Electric','Longroad Energy Services',
                'Los Vientos Windpower','Madison','Massachusetts','Meadow Lake Wind Farm','Mesquite','Michigan Wind',
                'Musselshell','NaturEner','NextEra','NorthWestern Energy','NRG','OEE','Osceola','Paulding Wind Farm','Pecos',
                'Pinyon Pines','Power County Wind Park','Rattlesnake','Rising Tree Wind Farm','Rock County']

def replace_similar_values(value):
    for standard_name in company_list:
        if process.extractOne(value, [standard_name])[1] > 80:
            return standard_name
    return value
```

```python
# function call
df['Operator Name'] = df['Operator Name'].apply(replace_similar_values)
```

#### We have the 2 dataframes now loaded in Python. we can fill the Null EIA_id with matching values from EIA 2017-2021 file by using Numpy.where function. We can also use pd.merge function via left and right join, but its best to keep the 2 dataframes separate to be loaded into Tableau. Later join can be created in Tableau using eia_id = Plant Id.

#### First we need to create a copy of df as df_eia and remove duplicate plant names in df_wind

```python
df_eia = df.copy()
```

```python
df_eia = df_eia.reset_index()
```

#### Lets filter df_wind by null eia_ids and check if there are duplicate plant name within the nulls

```python
null_eia_id = df_wind[df_wind['eia_id'].isnull()]

duplicate_plant_names = null_eia_id[null_eia_id.duplicated(subset='plant_name', keep=False)]
```

```python
print(null_eia_id['plant_name'].value_counts())
```

#### Now lets lookup eia_id from df_eia and use it to fill Null values in df_wind

```python
import numpy as np
```

```python
# Remove duplicates from the 'Plant Name' column in df
df_eia = df.drop_duplicates(subset='Plant Name', keep = False)


# Update values in df_wind.eia_id based on the condition
df_wind['eia_id'] = np.where(df_wind['eia_id'].isnull(),
    df_wind['plant_name'].map(df_eia.set_index('Plant Name')['Plant Id']),
    df_wind['eia_id']
)
```
